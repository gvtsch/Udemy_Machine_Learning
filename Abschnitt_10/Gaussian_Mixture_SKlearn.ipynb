```python
import numpy as np
np.random.seed(42)
```

```python
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from plotting_GMM import plot_results
from dataset import generate_dataset
```

```python
x, y = generate_dataset(10_000, 10_000)
```

```python
plt.scatter(x[:, 0], x[:, 1], c = y)
plt.show()
```

```python
n_components = 2
gauss = GaussianMixture(
    n_components=n_components,
    covariance_type='diag')
gauss.fit(x)
```

```python
print(f"Modell converged: {gauss.converged_}")
```

```python
covs=gauss.covariances_
means=gauss.means_

cov1=np.diag(covs[0])
cov2=np.diag(covs[1])
```

```python
print(f"Covs der Hauptdiagnolen d端r die 1. Zufallsverteilung:\n {np.diag(covs[0])}")
print(f"Covs der Hauptdiagnolen d端r die 2. Zufallsverteilung:\n {np.diag(covs[1])}")
print(f"Means: \n{means}")
```

```python
plot_results(
    X=x,
    Y=y,
    Y_=gauss.predict(x),
    means=means,
    covariances=[cov1, cov2],
    index = 0,
    title='GM'
)
```
